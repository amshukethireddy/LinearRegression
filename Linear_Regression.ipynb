```python
'''
Linear Regression from Scratch
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


'''
plotting
inputs: x-independent variable
        y-dependent variable
        y_hat:predicted y value
output: plots the graphs(line and scatter) based on the given input
'''
def plot(x,y,y_hat):
   #plt.pause(0.0001)
   plt.scatter(x,y,color='red',label='scatter') #plotting data point
   plt.plot(x,y_hat,color='blue',label='best_fit')#plotting the straight using initial values of w and b
   plt.xlabel('Ads')
   plt.ylabel('Sales')
   plt.show()
   #plt.legend(['scatter','best_fit'])
   #plt.clf()


'''
For calculation y_hat value(y predicted/target variable)
inputs: x-independent varible(feature-tupple)
        w-slope/wights
        b-bias/y-intercept
ouput: returns y_hat value

'''
def predict(x,w,b):
  y_hat=(x*w)+b #y_pred=wx+b
  return y_hat





'''
to compute the cost function
inputs: x-independent varible(feature-tupple)
        y-dependent variable
        w-slope/wights
        b-bias/y-intercept
output: returns the cost
formula:mean squared error=((y_pred-y)^2)/2m
'''
def cost(x,y,w,b):
   m=len(y)  #Number of training exmaples
   y_hat=predict(x,w,b) #Compute the y_hat
   errors=y_hat-y  # Compute the errors
   mse=(1 / (2 * m)) * np.sum(errors ** 2)  # Compute the mean squared error
   print("MSE: ",mse)
   return mse


'''
To compute Gradient Descent
inputs: x-independent varible(feature-tupple)
        y-dependent variable
        w-slope/wights
        b-bias/y-intercept
        learning rate(alpha)
        itertions
output: gradient descent is computed

'''
def gradient(x,y,w,b,learning_rate,iterations):
  m=len(x)
  for i in range (iterations):
    y_hat=predict(x,w,b)
    dw=(1/m)*sum((x)*(y_hat-y))
    db=(1/m)*sum(y_hat-y)
    #updating weights
    w=w-(learning_rate*dw)
    b=b-(learning_rate*db)
    #print(w,b)
  plot(x,y,y_hat)
  return w,b




'''
to read the input values and print the output values
'''
def main():
  #reading data
  data= pd.read_csv("advertising.csv")
  x=data['Ads'].values
  y=data['Sales'].values
  w=1
  b=0
  learning_rate=0.00006
  iterations=1000000
  y_hat=predict(x,w,b)
  plot(x,y,y_hat)
  cost(x,y,w,b)
  w,b=gradient(x,y,w,b,learning_rate,iterations)
  print("w: ",w)
  print("b: ",b)
  a=predict(163,w,b)
  print("New Data(Predicted value):",a)
  cost(x,y,w,b)




if __name__ == "__main__" :
    main()



```
